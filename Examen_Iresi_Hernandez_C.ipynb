```python
class Humans():

    def __init__(self,legs,arms):
        self.legs = legs ## Aqui tienes que asignar la otra variable, son los argumentos
        self.arms = arms

    def pier(self):
        return self.legs + " son el tipo de pierna que tiene Iresi " ##Estos son los polimorfismos

    def braz(self):
        return self.arms + " son el tipo de brazos que tiene Mariana"
```

```python
import numpy as np

Lista_1= np.random.uniform(8,13,12) ##Aqui estamos mostrando 12 valores aleatorios en decimales que esten en el intervalo de 8-13
Lista_1
```

```python
import pandas as pd
diccionario={"Precios_muestra":Lista_1}
df=pd.DataFrame(diccionario, index=["Mes 1", "Mes 2", "Mes 3", "Mes 4", "Mes 5", "Mes 6", "Mes 7", "Mes 8", "Mes 9", "Mes 10", "Mes 11", "Mes 12"])
df ##Creamos nuestro dataframe a partir de nuestra lista de numeros random y los asignamos a los meses del año
```

```python
##Calculo de los rendimietos simples
rendimientos=df.pct_change().dropna()
rendimientos
```

```python
media= df.mean()
media ##Esta es la media de todos los precios
```

```python
std = df.std()
std ##Esta es la volatilidad del portafolio de una forma mas sencilla
```
