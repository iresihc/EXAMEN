```python
##4.-
import pandas as pd
data=pd.DataFrame({'Country':['Russia','Colombia','Chile','Equador','Nigeria'],'Rank':[121,40,100,130,11]})
data
```

```python
#5.- Aqui esta devolviendo la informacion del DataFrame; el numero de filas, columnas, los indices y la memoria usada en éste
data.info()
```

```python
#6.-
# a) Las operaciones que se estan haciendo con los valores de las llaves es que se esta asignando el nombre de cada columna con las k, en k1 los valores tipos string de cada renglon y con el numero las veces que este nombre caracter se repetira dentro del dataframe hacia abajo, ademas de los valores tipo int que estos tendran dentro de k2
# b)
data = pd.DataFrame({'k1':['one']*3 + ['two']*4, 'k2':[3,2,1,3,3,4,4]})
data
```

```python
#7.- La funcion drop.duplicate tiene como funcion eliminar las columnas que se repitan dentro del DatafRAME
data.drop_duplicates(inplace=True)
data
```

```python
##8.- El siguiente Data Frame nos muestra valores unicamente que se encuentren dentro del 0-11 debido al arange impuesto, y como es 12 y empieza a contar desde el cero, se corta en el #11; con reshape indica el numero de filas y columnas que la tabla tendra, ademas de indicar con index los nombres de estas filas y con columns, lo mismo
data = pd.DataFrame(np.arange(12).reshape((3, 4)),index=['CDMX', 'Guanajuato', 'Puebla'],columns=['uno', 'dos', 'tres', 'cuatro'])
data
```

```python
##9.-
data = pd.DataFrame(np.arange(12).reshape((3, 4)),index=['UP', 'TEC', 'ANAHUAC'],columns=['uno', 'dos', 'tres', 'cuatro'])
data
```
